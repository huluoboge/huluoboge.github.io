第一课： numpy 的基本用法入门

```python
import numpy as np


points = np.array([
    [0,0,0],
    [1,0,0],
    [0,1,0],
    [0,0,1]
])

print(points)
print(points.shape)

print(points[:,2])
print(points[0:2,1])
```

```python
random_points = np.random.randn(1000,3)
print(random_points)
```

```python
trans_points = random_points + np.array([1,2,3])
print(trans_points)
```

```python
theta = np.pi/2
R = np.array(
    [
        [np.cos(theta), -np.sin(theta),0],
        [np.sin(theta), np.cos(theta),0],
        [0,0,1]
    ]
)
print(R)
print(R.shape)
```

```python
t = np.array([1,2,3])
print(t)
print(t.shape)
```

```python
transform_points = trans_points @ R.T + t
print(transform_points)
```

```python
v1 = np.array([1,2,3])
v2 = np.array([4,5,6])
dot_product = np.dot(v1,v2)
print(dot_product)
```

```python
cross_product = np.cross(v1, v2)
print(cross_product)
```

```python
norm_v1 = np.linalg.norm(v1)
norm_v2 = np.linalg.norm(v2)
print(norm_v1, norm_v2)
norm_l1_v1 = np.linalg.norm(v1, ord = 1)
norm_l2_v1 = np.linalg.norm(v1, ord = 2)
print(norm_l1_v1,norm_l2_v1)
```

```python
cos_theta = np.dot(v1,v2) / (np.linalg.norm(v1)*np.linalg.norm(v2))
print(cos_theta)
theta_rad = np.arccos(cos_theta)
print(theta_rad)
theta_deg = np.degrees(theta_rad)
```
