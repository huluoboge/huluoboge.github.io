ExitStack å¯ä»¥åˆ›å»ºä¸€ä¸ªä¸Šä¸‹æ–‡ç®¡ç†å™¨ï¼Œç”¨äºç®¡ç†å¤šä¸ªä¸Šä¸‹æ–‡ç®¡ç†å™¨ï¼Œå¹¶åœ¨é€€å‡ºæ—¶ç¡®ä¿å®ƒä»¬æŒ‰æ­£ç¡®çš„é¡ºåºé€€å‡ºã€‚
ç›¸å½“ä¸å¤šä¸ªwith



````python
from contextlib import ExitStack

from pathlib import Path

files = [f for f in Path.cwd().glob("*.ipynb") if f.is_file()]
# files = ["file1.txt", "file2.txt", "file3.txt"]
with ExitStack() as stack:
    fs = [stack.enter_context(open(f)) for f in files]
    # fs æ˜¯æ‰€æœ‰æ‰“å¼€çš„æ–‡ä»¶å¯¹è±¡
    for f in fs:
        print(f.read())

# æ‰€æœ‰æ–‡ä»¶éƒ½è‡ªåŠ¨å…³é—­
````
