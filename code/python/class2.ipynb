# 动手练习：批量法向量计算

## 1️⃣ 目标

给 N 个三角形的顶点 `v0, v1, v2`，计算每个三角形的单位法向量。



```python
import numpy as np
npoints = 1000
v0 = np.random.rand(npoints,3)
v1 = np.random.rand(npoints,3)
v2 = np.random.rand(npoints,3)
print(v0)
print(v1)
print(v2)
```

```python
v01 = v1-v0
v02 = v2-v0
face_normal = np.cross(v01,v02)
print(face_normal)
```

```python
normal = face_normal / np.linalg.norm(face_normal, axis=1,keepdims=True)
print(normal)
```
