NumPy 版本不兼容 

你当前安装的是 NumPy 2.0.2

但是某些已经编译的模块（比如某些 C++ 扩展或者 pybind11 模块）是基于 NumPy 1.x 编译的

直接用 NumPy 2 会崩溃，所以提示你要么 降级 NumPy 到 1.x，要么 升级受影响的模块 并重新编译

```python
img = torch.randn(1, 3, 256, 256).cuda()  # 1张3通道256x256图像
```

```python
import matplotlib.pyplot as plt

tensor_cpu = img[0].permute(1, 2, 0).cpu().numpy()  # 转成 HWC, CPU numpy
plt.imshow(tensor_cpu)
plt.show()
```
