远程 GPU Notebook 配置流程
需要远程安装gpu的内核
python -m ipykernel install --user --name gpu --display-name "Python (GPU)"

远程端口映射
ssh -L 8888:localhost:8888 recon@10.33.207.203

远程服务上开启,--NotebookApp.token=''表示不用密码

jupyter notebook --no-browser --port=8888 --NotebookApp.token=''
